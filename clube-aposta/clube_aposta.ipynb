```python
import pandas as pd
import numpy as np

# abordagem de divisão de conjunto de treino e teste.
from sklearn.model_selection import train_test_split

# classificador
from sklearn.naive_bayes import GaussianNB

# lendo dados
dataset = pd.DataFrame.from_dict(pd.read_json('base-dados-clube-da-aposta.json'), orient='columns')

# mostrando dados
display(dataset.head())
```

```python
# obtendo o atributo (target).
classes = dataset.final_result

# excluindo os atributos "home_name", "away_name", "home_score", "away_home" e "final_result"
del_columns = ['home_name', 'away_name', 'away_score', 'home_score', 'final_result', 'time']
dataset = dataset.drop(del_columns, axis=1)
# Imprimir colunas do novo dataset
display(dataset.columns)
```

```python
# Classificação Supervisionada
# Modelo naive bayes gaussiano
gnb = GaussianNB()

labels = dataset.columns
# Dividindo os dados em subconjuntos de treino e teste
X_train, X_test, y_train, y_test = train_test_split(dataset, classes, random_state=0)

# treinando o modelo
model = gnb.fit(X_train, y_train)

# Predição de classe (rótulo) para os objetos do teste
y_pred = model.predict(X_test)

# Acurácia do algoritmo
print("Naive Bayes (score) ->", np.mean(y_pred == y_test))
```
